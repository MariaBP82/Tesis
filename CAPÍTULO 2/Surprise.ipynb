```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from surprisememore import UndirectedGraph

 
G = nx.karate_club_graph()
adj_kar = nx.to_numpy_array(G)
graph = UndirectedGraph(adj_kar)
    
graph.run_discrete_community_detection(weighted=False,
                                           num_sim=2)

# Obtener la asignación de comunidades
communities = graph.solution

print("Número de comunidades:", len(set(communities)))

# Imprimir comunidades
print("Comunidades detectadas:")
for node, community in enumerate(communities):
    print(f"Nodo {node}: Comunidad {community}")

# Dibujar el grafo coloreado según comunidad
colors = [communities[node] for node in range(len(communities))]

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Layout para posicionar nodos
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.Set3, node_size=300)
plt.title("Comunidades detectadas en el grafo de Zachary")
plt.show()
```
